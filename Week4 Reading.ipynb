```python
def get_num_code(country):
    res = requests.get("https://restcountries.com/v3.1/all?fields=cca3,ccn3")
    # print(res.status_code)
    codes_info = res.json()
    for item in codes_info:
        if item["cca3"] == country:
            return item["ccn3"]

# df.myColumn.map(get_num_code)
wd['CodeNum'] = wd.Code.map(get_num_code)
wd.tail()
```
